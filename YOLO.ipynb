```python
# ============================================================
# CREATE data.yaml
# ============================================================
import yaml, os

DETECT_PATH = r'D:\Tejashri\Aeriel\Object_detection_dataset'

data_yaml = {
    'path'  : DETECT_PATH,
    'train' : 'train/images',
    'val'   : 'valid/images',
    'test'  : 'test/images',
    'nc'    : 2,
    'names' : {0: 'bird', 1: 'drone'}
}

yaml_path = os.path.join(DETECT_PATH, 'data.yaml')
with open(yaml_path, 'w') as f:
    yaml.dump(data_yaml, f, default_flow_style=False, sort_keys=False)

print(f"✅ data.yaml saved: {yaml_path}")
print("\nContents:")
with open(yaml_path, 'r') as f:
    print(f.read())
```

```python
# ============================================================
# TRAIN YOLOv8
# ============================================================
from ultralytics import YOLO

model_yolo = YOLO('yolov8n.pt')    # nano — lightest for CPU

model_yolo.train(
    data=yaml_path,
    epochs=50,
    imgsz=640,
    batch=8,
    name='bird_drone_yolov8',
    project='yolo_runs',
    patience=10,
    device='cpu',
    verbose=True
)

print("\n✅ Training complete")
print("   Best weights: yolo_runs/bird_drone_yolov8/weights/best.pt")
```
